## Introduction


**Gun safety is of valuable concern to the United States security. For the reason of personal security/hunting purposes, it is possible for individual to own guns. To keep a track of the number of buyers, and also to be informed of the backkground of the prospective client is of utmost importance for gun dealers.**

**Hence to keep track of this, FBI's National Instant Criminal Background Check System (NICS) has maintained a database of the background of buyers. The NICS is hence used by dealers to determine whether a prospective buyer is eligible to buy firearms or explosives. Gun shops call into this system to ensure that each customer does not have a criminal record or isn’t otherwise ineligible to make a purchase. Further, there is also a record on the number of buyers across states within the country.**

**In this project, we will use the dataset (firearms checks made by NICS) to determine and analyse trends in the firearms buyers in the US with respect to the states,in particular address the following questions:**
<br>
**1. The states with highest gun purchasers. **
<br>
**2. The state with highest total gun purchaser/rental and its composition. **
<br>
**3. The overall trend of the United States gun purchases for the amount of data available. **

## Exploratory Data Analysis



### Analysing the number of handguns, long guns and the total number of purchases 

**1. At first, we will look at the trend in the number of gun purchases over the last 20 years, beginning 1998 until 2017. **

```python
"""Plotting the average number of gun purchases over the years"""

plt.figure(figsize=(15,14))
df1.groupby(['month'])['totals'].mean().plot()

"""Getting the xtick positions"""
x=[]
for i in np.arange(3,228,12):
  x.append(i)

"""Getting the xtick labels that correspond to the year of gun purchase"""
label_x=[]
for i in np.arange(1999,2018,1):
  label_x.append(i)
    

plt.xticks(x, label_x, rotation='vertical')
plt.title('Total number of gun purchaces across the states',fontsize=14)
plt.ylabel('Average number of gun purchases in a month',fontsize=14)
plt.xlabel('Year',fontsize=14);
```

** The above plot shows us that on an average there has been an increasing trend in the number of gun purchases starting from the year 1998 while there has been significant fluctuations within the different months of the year. The average has increased from 1000 purchases in 1999 to close to 50,000 in 2017.**

**2. Despite having a variety of gun, we will here restrict ourselves with handguns and long guns. At first, we will have a look at the number of handgun purchases by state. **

```python
"""Plotting the average number of handgun purchases for different states"""
plt.figure(figsize=(15,8))
df1.groupby(['state'])['handgun'].mean().plot(kind='bar')
plt.title('Number of handguns',fontsize=14)
plt.ylabel('Average number of handguns per state',fontsize=14)
plt.xlabel('State',fontsize=14);
```

** The number of hand guns purchases on aan average are the largest in California, Florida and Texas.**

** Secondly, we will have a look at the number of long gun purchases by state. **

```python
"""Plotting the average number of long gun purchases for different states"""
plt.figure(figsize=(15,8))
df1.groupby(['state'])['long_gun'].mean().plot(kind='bar')

plt.title('Number of long guns',fontsize=14)
plt.ylabel('Average number of long guns per state',fontsize=14)
plt.xlabel('State',fontsize=14);
```

**Here, we will have a look at the number of total number of gun purchases by state.**

```python
"""Plotting the average number of total gun purchases/rentals for different states"""
plt.figure(figsize=(15,8))


df1.groupby(['state'])['totals'].mean().plot(kind='bar')

plt.title('Total number of guns',fontsize=14)
plt.ylabel('Average number of guns',fontsize=14)
plt.xlabel('State',fontsize=14);
```

** The highest gun purchaser is Kentucky while the Texas, California, Florida and Pennsylvania are also moderately heavy buyers.**

** Another thing to notice is that, there is no one-to-one correlation between the largest hand gun/long gun purchasing state and the overall large purchaser since we have neglected the other types of guns in the sample considered. **

** After having identified the largest buyer, Kentucky, if not the largest in long gun/handgun, we will look at its subtler aspects in terms of the types of handgun/long gun that has been purchased by it. **
** Since there are several null values before 2015 for the other columns, we will use the trimmed dataset 'after_2015' to look at the split for Kentucky.**

```python
"""Identifying the split up of different parameters for the state with largest gun purchaser/rental"""
after_2015.loc[after_2015['totals'].idxmax()]
```

```python
"""Identifying the distribution of the permits of the state with largest gun purchaser/rentals"""

plt.figure(figsize=(15,8))


df1.permit[df1['state'] == 'Kentucky'].hist(alpha=0.5,label='Kentucky')

plt.title('Distribution of total number of permits provided',fontsize=16)
plt.xlabel('Total number of guns',fontsize=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.legend(fontsize =16);
```

**Finally, looking at the histogram of the total gun purchases of some of the highest gun purchasers (California, Kentucky, Texas, Florida and Pennsylvania) as shown below, we realize all of the states are skewed to the left. **
**This plot probably also gives us a hint that despite not being the highest in handgun/long gun purchases, Kentucky has a wider distribution of purchases with a higher number over the years compared to the other states. Hence, its overall average purchase has been highest. While Texas and California decay slowly, Pennsylvania decays much faster indicating its concentration within one particular range.**

```python
"""Identifying the distribution of the state with largest gun purchaser/rentals"""

plt.figure(figsize=(15,8))

df1.totals[df1['state'] == 'California'].hist(alpha=0.5,label='California')
df1.totals[df1['state'] == 'Kentucky'].hist(alpha=0.5,label='Kentucky')
df1.totals[df1['state'] == 'Florida'].hist(alpha=0.5,label='Florida')
df1.totals[df1['state'] == 'Pennsylvania'].hist(alpha=0.5,label='Pennsylvania')
df1.totals[df1['state'] == 'Texas'].hist(alpha=0.5,label='Texas')


plt.title('Distribution of total number of guns purchased/rented',fontsize=16)
plt.xlabel('Total number of guns',fontsize=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.legend(fontsize =16);
```
